## Plotting in python

We are going to use the matplotlib package to plot

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 100)
```

```python
plt.plot(x)
plt.show()
```

It assumed our single data list to be the y-values and in the absence of an x-values list, [0, 1, 2, 3, ... 100] was used instead

```python
y = np.arange(0,1,0.01)
```

```python
plt.plot(x,y)
plt.show()
```

Now we are passing in two lists to plt.plot, so you are explicitly setting the x-values. 

Very important: When you pass in two lists, their lengths must match

```python
y1 = np.arange(0,1,0.01)
y2 = np.arange(2,3,0.01)
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

You can plot multiple curves on same plot, but they have to all be the same length.

You cannot tell which line is for which dataset. You can add labels and show a legend.

```python
plt.plot(x,y1,label="plot 1")
plt.plot(x,y2,label="plot 2")
plt.legend()
plt.show()
```

Add labels to your x and y axis as well as a title to make the plot more informative

```python
plt.plot(x,y1,label="plot 1")
plt.plot(x,y2,label="plot 2")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a plot of two datasets")
plt.legend()
plt.show()
```

You can change the line types as well is you want a dashed line. 
You can also set the color of the line. 

```python
plt.plot(x,y1,linestyle=":",label="plot 1")
plt.plot(x,y2,color="red",label="plot 2")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a plot of two datasets")
plt.legend()
plt.show()
```

### Other types of plots in matplotlib

```python
# Scatter plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a scatter plot")
plt.show()
```

```python
# Boxplot
data = np.random.rand(50)
plt.boxplot(data)
plt.ylabel("Y-label")
plt.title("This is a boxplot")
plt.show()
```

```python
# Multiple Boxplot
data1 = np.random.rand(50)
data2 = np.random.rand(50)
# Make a list of the two datasets
data = [data1,data2]
plt.boxplot(data)
plt.ylabel("Y-label")
plt.title("This is a boxplot with two datasets")
plt.show()
```

```python
# Histogram
data = np.random.rand(10000)
plt.hist(data, bins=50, facecolor='green')
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a histogram")
plt.show()
```

```python
# Histogram which is a probability density
data = np.random.rand(10000)
plt.hist(data, bins=50, normed=1, facecolor='green')
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a probability density")
plt.show()
```

What if we want to have multiple subplots within the same figure? 

Let's combine our scatterplot, boxplot and histogram into one figure.

Use the command plt.subplot(numrows,numcols,plotNum) 

```python
plt.figure(figsize=(16, 12))
plt.subplot(3, 1, 1)
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.ylabel("Y-label")
plt.title("This is a scatter plot")

plt.subplot(3, 1, 2)
data = np.random.rand(50)
plt.boxplot(data)
plt.ylabel("Y-label")
plt.title("This is a boxplot")

plt.subplot(3, 1, 3)
data = np.random.rand(10000)
plt.hist(data, bins=50, facecolor='green')
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.title("This is a histogram")

plt.show()
```
