# Modules

- Packages of functions that we can load into memory at the start of a script
- import _moduleName_ as _abbreviation_

```python
import random as rn

for x in range(10):
    print(rn.randint(0,10))
```

- Instead of loading the entire module, a specific function or functions can be loaded instead
- the function seed() just allows our supposedly "random" number generator to give the same results each run

```python
from random import randint,seed
seed(20)
age = []
survived = []
for x in range(100):
    age.append(randint(20,70))
    survived.append(randint(0,1))
```

```python
print(age)
print(survived)
```

# Matplotlib (pyplot) vs ggplot2

- matplotlib is the plotting library for python (and matlab)
- ggplot2 is the plotting library for R
- they have different approaches for creating a figure and are not easily translated between one another
- we will be focusing on using matplotlib

```python
import matplotlib.pyplot as plt
#next line just lets plots appear inline; ipython terminal specific
%matplotlib inline 
```

Creating figures in matplot take the following approach:

1) create your canvas (your figure)

2) add layers to your canvas (each being applied on top of the previous)

3) show your canvas

Layers can take the form of anything from full histogram graphs to single dots and lines


```python
#Create canvas
plt.figure()

#Add Layers
plt.hist(age)               #histogram graph
plt.axhline(4,color='r')    #horizontal line
plt.axvline(50,color='y')   #vertical line

#Show canvas
plt.show()
```

# Seaborn Module 

Matplotlib can get very confusing to use and can get quite messy, so the seaborn module was developed as a high-level (i.e. easier to interpret/use) interface to using matplotlib

```python
import seaborn as sns
```

```python
#Create Canvas
sns.plt.figure()

#Layers
sns.distplot(age)

#Show
sns.plt.show()
```

seaborn comes with many default styles

```python
sns.set_style('white')
sns.set_context('talk')

sns.plt.figure()
sns.distplot(age)
sns.plt.show()
```

seaborn also has very nice documentation that describes all the different ways you can change a plot through parameters
http://seaborn.pydata.org/generated/seaborn.distplot.html

```python
sns.set_style("darkgrid")
sns.plt.figure()
sns.distplot(age, rug=True, rug_kws={"color": "g"}, kde_kws={"color": "k", "lw": 3, "label": "KDE"}, hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "g"})
sns.plt.show()
```

# Plotting

#### Seaborn is nicely integrated with pandas and makes plotting with dataframes very intuitive

```python
#Read in entire csv file as a dataframe
df = pd.read_csv('Pokemon.csv',index_col=0)
```

```python
df.head()
```

```python
sns.set_style('darkgrid')
sns.set_context('talk')

sns.plt.figure(figsize=(10,6))
#Select the dataframe to pull data from, then specify column names for x and y
sns.lmplot(data=df,x='Attack',y='Defense')
sns.plt.show()
```

#### There is no "scatterplot" in seaborn.  However, you can easily use lmplot and turn off the regression line.

In addition, we can tell seaborn to color our points using another column in the dataframe

```python
sns.set_context('notebook')
sns.plt.figure(figsize=(10,6))
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False,hue='Stage')
sns.plt.show()
```

#### Seaborn is matplotlib at its core so we can use matplotlib commands to further customize our figure

```python
sns.set_context('notebook')
sns.plt.figure(figsize=(10,6))
sns.lmplot(data=df,x='Attack',y='Defense',fit_reg=False,hue='Stage')
sns.plt.axvline(60)
sns.plt.show()
```

#### Seaborn will try to interpret what you ask.

Just calling boxplot here will create a boxplot for each column in your dataframe

```python
sns.plt.figure(figsize=(10,6))
sns.boxplot(data=df)
sns.plt.show()
```

 To fix this, we use pandas to create a subset dataframe from our original

```python
sns.set_style("whitegrid")
stats_df = df.drop(['Total','Stage','Legendary'], axis=1)
sns.boxplot(data=stats_df)
```

#### Violinplots

```python
sns.violinplot(data=df, x='Type 1', y='Attack')
```

#### Seaborn has an extensive color options
- http://seaborn.pydata.org/tutorial/color_palettes.html

Here, we use our own manually defined colors in hexidecimal
- https://color.adobe.com/

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
sns.set_style("whitegrid")

sns.plt.figure()
sns.violinplot(data=df, x='Type 1', y='Attack',palette=pkmn_type_colors)
sns.plt.show()
```

#### Increase the size of the figure to prevent crowding
- Define the figure size when 'creating' your canvas

```python
sns.plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Type 1', y='Attack',palette=pkmn_type_colors)
sns.plt.show()
```

#### Swarmplot

```python
sns.plt.figure(figsize=(10,6))
sns.swarmplot(data=df, x='Type 1', y='Attack', palette=pkmn_type_colors)
sns.plt.show()
```

Seaborn plots are simple layers that can be combined

```python
sns.plt.figure(figsize=(10,6))
sns.violinplot(data=df, x='Type 1', y='Attack',palette=pkmn_type_colors,inner=None)
sns.swarmplot(data=df, x='Type 1', y='Attack', color='black')
sns.plt.show()
```

### Advanced Plotting

```python
stats_df.head()
```

```python
melted_df = pd.melt(stats_df,
                    id_vars=["Name", "Type 1", "Type 2"],
                    var_name="Stat")
sns.plt.figure(figsize=(10,6))
sns.swarmplot(data=melted_df, x='Stat', y='value', hue='Type 1',palette=pkmn_type_colors,split=True)
sns.plt.legend(bbox_to_anchor=(1,1),loc=2)
sns.plt.ylim(0)
sns.plt.show()
```

# Extra

Finding pokemon with highest defense

Finding pokemon with 'best attack and defense' (weighing attack)

```python
df[['Defense','Attack']].head()
```

Someone asked if you could could the "best" pokemon using weighted stats.

```python
def calculateImportance(row):
    return int(row['Defense']) + int(row['Attack']*2)

df.loc[df[['Defense','Attack']].apply(lambda row: calculateImportance(row),axis=1).argmax()]
```
